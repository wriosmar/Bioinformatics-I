```python
# A function that takes the reverse complement of a given text.
def reverse_complement(text):
    reverse = text[::-1]
    complement = ""

    for i in reverse:
        if i == "A":
            char = "T"
        if i == "T":
            char = "A"
        if i == "C":
            char = "G"
        if i == "G":
            char = "C"
        complement = complement + char

    return complement
```
